```python
def gcd(x,y):
        
    if x>y:
        x,y = y,x
        
    if x ==0:
        
        return y
    
    else:
        
        return gcd(y%x,x)
```

```python
print(gcd(0,10))
```

```python
print(gcd(10,0))
```

```python
print(gcd(1,10))
```

```python
print(gcd(10,1))
```

```python
print(gcd(2,10))
```
